```python
import pandas as pd
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
# main directory
path = "/Users/ritvikraina/Library/Mobile Documents/com~apple~CloudDocs/College/Courses/CS4465 Bioinformatics/Project/data/dataset3"
```

```python
# Create train and test directories
train_path = os.path.join(path,'train')
test_path = os.path.join(path,'test')
```

```python
#sub-directories
covid_train_path = os.path.join(train_path,'Covid')
normal_train_path = os.path.join(train_path,'Normal')


covid_test_path = os.path.join(test_path,'Covid')
normal_test_path = os.path.join(test_path,'Normal')
```

```python
covid_df = pd.read_csv(os.path.join(path,"meta_data_covid.csv"),encoding="windows-1252")
normal_df = pd.read_csv(os.path.join(path,"meta_data_normal.csv"),encoding="windows-1252")
```

```python
covid_df.drop(['Patient ID', 'Gender', 'Age', 'Country','Slice','Case Discussion', 'Case Number', 'Columns',
       'Date', 'Date of Last Calibration', 'Distance Source to Detector',
       'Distance Source to Patient', 'Exposure Time', 'Image Source',
       'Institution', 'KVP', 'Last Calibration', 'Link', 'Link2',
       'Other diseases', 'Presentation', 'RT_PCR_COVID-19', 'Rows', 'Severity',
       'Slice Thickness', 'Slices', 'Source_detail', 'Time', 'XRayTubeCurrent',
       'category', 'doi', 'finding', 'in_icu', 'intubated',
       'intubation_present', 'license', 'mAs', 'mask_file',
       'needed_supplemental_O2', 'study_file', 'survival', 'went_icu'],axis=1,inplace=True)
```

```python
normal_df.drop(['index', 'Patient ID', 'Gender', 'Age', 'Country', 'Slice', 'Case Discussion', 'Case Number', 'Columns',
       'Date', 'Date of Last Calibration', 'Distance Source to Detector',
       'Distance Source to Patient', 'Exposure Time', 'Image Source',
       'Institution', 'KVP', 'Last Calibration', 'Link', 'Link2',
       'Other diseases', 'Presentation', 'RT_PCR_positive', 'Rows', 'Severity',
       'Slice Thickness', 'Slices', 'Source_detail', 'Time', 'XRayTubeCurrent',
       'category', 'doi', 'finding', 'in_icu', 'intubated',
       'intubation_present', 'license', 'mAs', 'mask_file',
       'needed_supplemental_O2', 'study_file', 'survival', 'went_icu'],axis=1,inplace=True)
```

```python
normal_df
```

```python
df = pd.concat([normal_df,covid_df],ignore_index=True)
df
```

```python
# Images and Labels
x = df.loc[:,'File name']
y = df.loc[:,'Diagnosis']

# Train-Test splitfor train and validation images
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, stratify=y)
```

```python
# Train df
df_train = pd.DataFrame(columns=['Image','Diagnosis'])
df_train['Image'] = x_train
df_train['Diagnosis'] = y_train

# Test df
df_test= pd.DataFrame(columns=['Image','Diagnosis'])
df_test['Image'] = x_test
df_test['Diagnosis'] = y_test

# Reset index
df_train.reset_index(drop=True, inplace=True)
df_test.reset_index(drop=True, inplace=True)
```

```python
# Save train images
for i in range(len(df_train)):

    image = df_train.loc[i,'Image']

    if df_train.loc[i,'Diagnosis'] == "Normal":
        shutil.copy(path + '/curated_data/Normal/' + image, normal_train_path)
    else:
        shutil.copy(path + '/curated_data/Covid/' + image, covid_train_path)
```

```python
# Save test images
for i in range(len(df_test)):

    image = df_test.loc[i,'Image']

    if df_test.loc[i,'Diagnosis'] == "Normal":
        shutil.copy(path + '/curated_data/Normal/' + image, normal_test_path)
    else:
        shutil.copy(path + '/curated_data/Covid/' + image, covid_test_path)
```
